```
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
wordcloud = WordCloud(
                      stopwords=STOPWORDS,
                      background_color='white',
                      width=1200,
                      height=1000
                     ).generate(text)


plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```
plt.savefig('Capitalism.png')
```

```
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
wordcloud = WordCloud(
                      stopwords=STOPWORDS,
                      background_color='black',
                      width=2000,
                      height=1200
                     ).generate(text2)


plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```
plt.savefig('Digital Health.png')
```
